```python
from bokeh.models import FuncTickFormatter

def boxplot_plot(vegan, baseline):
    # graph labels and positions 
    moods = ['Cheerful', 'Insecure', 'Relaxed', 'Irritated', 'Satisfied',
                    'Down', 'Energy', 'Anxiety', 'Happy', 'Lonely', 'Concentration',
                    'Worried', 'Depression', 'Comfortability', 'Active', 'Hunger', 
                    'Tired', 'Pain', 'General feeling']

    base_positions = [0.5, 2.5, 4.5, 6.5, 8.5, 10.5, 12.5, 14.5, 16.5, 18.5, 20.5, 22.5 ,24.5 ,26.5 ,28.5 ,30.5 ,32.5 ,34.5 ,36.5]

    vegan_positions = [pos+0.6 for pos in base_positions]
    label_positions = [pos+0.3 for pos in base_positions]
    graph_labels = str(dict(zip(label_positions, moods)))
    outlier_plot_positions_b = dict(zip(moods, base_positions))
    outlier_plot_positions_v = dict(zip(moods, vegan_positions))

    # find the quartiles and IQR for each category
    groups_b = baseline.groupby('mood')
    groups_v = vegan.groupby('mood')

    q1_b = groups_b.quantile(q=0.25)
    q1_v = groups_v.quantile(q=0.25)

    q2_b = groups_b.quantile(q=0.5)
    q2_v = groups_v.quantile(q=0.5)

    q3_b = groups_b.quantile(q=0.75)
    q3_v = groups_v.quantile(q=0.75)

    iqr_b = q3_b - q1_b
    iqr_v = q3_v - q1_v

    upper_b = q3_b + 1.5*iqr_b
    upper_v = q3_v + 1.5*iqr_v

    lower_b = q1_b - 1.5*iqr_b
    lower_v = q1_v - 1.5*iqr_v

    out_b = groups_b.apply(get_outliers, upper=upper_b, lower=lower_b).dropna()
    out_v = groups_v.apply(get_outliers, upper=upper_v, lower=lower_v).dropna()

    # prepare outlier data for plotting, we need coordinates for every outlier.
    if not out_b.empty:
        outx_b = []
        outy_b = []
        for keys in out_b.index:
            outx_b.append(keys[0])
            outy_b.append(out_b.loc[keys[0]].loc[keys[1]])

    if not out_v.empty:
        outx_v = []
        outy_v = []
        for keys in out_v.index:
            outx_v.append(keys[0])
            outy_v.append(out_v.loc[keys[0]].loc[keys[1]])

    # p = figure(plot_width = 1000, x_range=(moods), y_range=(0, 8), title="Boxplots representing mental health moods")
    p = figure(tools="", plot_width=900, y_range=(0.5, 7.5), x_axis_label="Mood", y_axis_label="Score")

    # if no outliers, shrink lengths of stems to be no longer than the minimums or maximums
    qmin_b = groups_b.quantile(q=0.00)
    qmin_v = groups_v.quantile(q=0.00)

    qmax_b = groups_b.quantile(q=1.00)
    qmax_v = groups_v.quantile(q=1.00)

    upper_b.score = [min([x,y]) for (x,y) in zip(list(qmax_b.loc[:,'score']),upper_b.score)]
    upper_v.score = [min([x,y]) for (x,y) in zip(list(qmax_v.loc[:,'score']),upper_v.score)]

    lower_b.score = [max([x,y]) for (x,y) in zip(list(qmin_b.loc[:,'score']),lower_b.score)]
    lower_v.score = [max([x,y]) for (x,y) in zip(list(qmin_v.loc[:,'score']),lower_v.score)]

    # stems
    p.segment(base_positions, upper_b.score, base_positions, q3_b.score, line_color="black")
    p.segment(base_positions, lower_b.score, base_positions, q1_b.score, line_color="black")
    p.segment(vegan_positions, upper_v.score, vegan_positions, q3_v.score, line_color="black")
    p.segment(vegan_positions, lower_v.score, vegan_positions, q1_v.score, line_color="black")

    # boxes
    p.vbar(base_positions, 0.4, q2_b.score, q3_b.score, fill_color="#E08E79", line_color="black")
    p.vbar(base_positions, 0.4, q1_b.score, q2_b.score, fill_color="#3B8686", line_color="black")
    p.vbar(vegan_positions, 0.4, q2_v.score, q3_v.score, fill_color="#E08E79", line_color="black")
    p.vbar(vegan_positions, 0.4, q1_v.score, q2_v.score, fill_color="#3B8686", line_color="black")

    # whiskers (almost-0 height rects simpler than segments)
    p.rect(base_positions, lower_b.score, 0.2, 0.01, line_color="black")
    p.rect(base_positions, upper_b.score, 0.2, 0.01, line_color="black")
    p.rect(vegan_positions, lower_v.score, 0.2, 0.01, line_color="black")
    p.rect(vegan_positions, upper_v.score, 0.2, 0.01, line_color="black")

    # outliers
    if not out_b.empty:
        p.circle(outx_b, outy_b, size=6, color="#F38630", fill_alpha=0.6)

    if not out_v.empty:
        p.circle(outx_v, outy_v, size=6, color="#F38630", fill_alpha=0.6)


    p.xgrid.grid_line_color = None
    p.ygrid.grid_line_color = "white"
    p.grid.grid_line_width = 2
    p.xaxis.major_label_orientation = "vertical"

    # set labels at the tick positions
    p.xaxis.formatter = FuncTickFormatter(code="""var mapping = {};return mapping[tick];""".format(moods))
    
    show(p)
    

 # find the outliers for each category
def get_outliers(group, upper, lower):
    '''
    Function that finds the outliers for each category
    '''

    cat = group.name
    return group[(group.score > upper.loc[cat]['score']) | (group.score < lower.loc[cat]['score'])]['score']
```
