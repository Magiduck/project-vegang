```python
#The Columns to be reversed
cols = ['insecure', 'irritated', 'down', 'anxiety', 'lonely', 'worried', 'depression',
       'rather doing something else', 'hunger', 'tired', 'pain']

def reverseScoring(df, high, cols):
    df[cols] = high - df[cols]
    return df

def process(df):
    health_subject = reverseScoring(df, 7, cols)
    return health_subject
    
health_subject_a = process(health_subject_a)
health_subject_b = process(health_subject_b)
health_subject_c = process(health_subject_c)
health_subject_d = process(health_subject_d)
health_subject_e = process(health_subject_e)
```

```python
def vegan(df):
    vegan =  df[(df['date'].dt.day >= 26) | (df['date'].dt.month == 11)]
    return vegan

def baseline(df):
    baseline = df[(df['date'].dt.day <= 25) & (df['date'].dt.month == 10)]
    return baseline

vegan_a = vegan(health_subject_a)
baseline_a = baseline(health_subject_a)

vegan_b = vegan(health_subject_b)
baseline_b = baseline(health_subject_b)

vegan_c = vegan(health_subject_c)
baseline_c = baseline(health_subject_c)

vegan_d = vegan(health_subject_d)
baseline_d = baseline(health_subject_d)

vegan_e = vegan(health_subject_e)
baseline_e = baseline(health_subject_e)
```

```python
def average(df):
    average = df.mean().dropna().round(2)
    return average

health_subject_a_mean = average(health_subject_a)
baseline_a_mean = average(vegan_a)
vegan_a_mean = average(baseline_a)

health_subject_b_mean = average(health_subject_b)
baseline_b_mean = average(vegan_b)
vegan_b_mean = average(baseline_b)

health_subject_c_mean = average(health_subject_c)
baseline_c_mean = average(vegan_c)
vegan_c_mean = average(baseline_c)

health_subject_d_mean = average(health_subject_d)
baseline_d_mean = average(vegan_d)
vegan_d_mean = average(baseline_d)

health_subject_e_mean = average(health_subject_e)
baseline_e_mean = average(vegan_e)
vegan_e_mean = average(baseline_e)
```

```python
def reform(df):
    reformed = df.to_frame().reset_index().rename(columns={'index':'mood', 0 : 'score'})
    return reformed 

health_subject_a_mean = reform(health_subject_a_mean)
baseline_a_mean = reform(baseline_a_mean)
vegan_a_mean = reform(vegan_a_mean)

health_subject_b_mean = reform(health_subject_b_mean)
baseline_b_mean = reform(baseline_b_mean)
vegan_b_mean = reform(vegan_b_mean)

health_subject_c_mean = reform(health_subject_c_mean)
baseline_c_mean = reform(baseline_c_mean)
vegan_c_mean = reform(vegan_c_mean)

health_subject_d_mean = reform(health_subject_d_mean)
baseline_d_mean = reform(baseline_d_mean)
vegan_d_mean = reform(vegan_d_mean)

health_subject_e_mean = reform(health_subject_e_mean)
baseline_e_mean = reform(baseline_e_mean)
vegan_e_mean = reform(vegan_e_mean)
```
