```python
import yaml
from glob import glob
import pandas as pd
from IPython.display import display
from tqdm import tqdm
import yaml
import os
from ete3 import NCBITaxa
import numpy as np
from Bio import Entrez
import itertools
```

```python
def get_config():
    with open("config.yaml", 'r') as con:
        config = yaml.safe_load(con)
    return config

config = get_config()
taxa_profiles = pd.read_csv(config["taxa_profiles"], sep="\t")
taxa_profiles = taxa_profiles.set_index(["subject", "is_vegan"])
taxa_profiles = taxa_profiles.drop(['Unnamed: 0'], axis=1)
#taxa_profiles = taxa_profiles.drop("Unnamed: 0") 
```

```python
order_taxa = ['kingdom_id', 'phylum_id', 'class_id', 'order_id', 'family_id', 'genus_id', 'species_id']
x = taxa_profiles.to_numpy().tolist()
x = list(itertools.chain.from_iterable(x))
x = list(pd.unique(x))
x.remove(0)
Entrez.email ='m.a.de.la.rambelje@st.hanze.nl'
handle = Entrez.efetch('taxonomy', id=x)
response = Entrez.read(handle)
#print(response)
values = []
keys = []
for entry in response:
    values.append(entry.get("ScientificName"))
    keys.append(entry.get('TaxId'))
    
YES = dict(zip(keys, values))
print(YES)
```
